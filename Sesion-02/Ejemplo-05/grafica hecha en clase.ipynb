```
import numpy as np
import seaborn as sns
```

```
normal = []
for a,b in [[2,1],[-2,3],[0,2]]:
    print(a,b)
    normal.append(np.random.normal(loc=a,scale=b,size=100000))
```

```
for n in normal:
    sns.kdeplot(n,fill = True)
```

```
for n in normal:
    sns.kdeplot(n,fill = True, bw_adjust=.25)
```
